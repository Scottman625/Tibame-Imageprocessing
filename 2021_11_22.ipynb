```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=cv2.imread("1.jpeg", 1)
cv2_imshow(m1)
Bavg=m1[:,:,0].mean()
Gavg=m1[:,:,1].mean()
Ravg=m1[:,:,2].mean()
Cavg=(Bavg+Gavg+Ravg)
m1[:,:,0]=cv2.multiply(m1[:,:,0], (Cavg/(Bavg*3)))
m1[:,:,1]=cv2.multiply(m1[:,:,1], (Cavg/(Bavg*3)))
m1[:,:,2]=cv2.multiply(m1[:,:,2], (Cavg/(Bavg*3)))
cv2_imshow(m1)
```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=cv2.imread("1.jpg", 1)
m2=m1.copy()
m2[:,:,0]=cv2.adaptiveThreshold(m1[:,:,0],255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,55,3)
m2[:,:,0]=cv2.adaptiveThreshold(m1[:,:,1],255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,55,3)
m2[:,:,0]=cv2.adaptiveThreshold(m1[:,:,2],255, cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,55,3)
cv2_imshow(m1)
cv2_imshow(m2)
```
