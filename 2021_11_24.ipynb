```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow

m1=cv2.imread("3.jpg", 1)


m2=cv2.inRange(m1, (0,0,0), (100,100,100))
m2=cv2.dilate(m2, np.ones((9,14)))
a, b=cv2.findContours(m2,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
# cv2.drawContours(m1,a,-1,(200,0,0),2)
hc=int(m1.shape[0]*0.5)
wc=int(m1.shape[1]*0.5)
centerIndex=0
for i in range(0,len(a),1):
  x1,y1,w1,h1=cv2.boundingRect(a[centerIndex])
  r1cx=int(x1+(w1*0.5))
  r1cy=int(y1+(h1*0.5))

  x2,y2,w2,h2=cv2.boundingRect(a[i])
  r2cx=int(x2+(w2*0.5))
  r2cy=int(y2+(h2*0.5))
  if abs(r1cx-wc)+abs(r1cy-hc)>abs(r2cx-wc)+abs(r2cy-hc):
    centerIndex=i
x,y,w,h=cv2.boundingRect(a[centerIndex])
cv2.rectangle(m1, (x,y), (x+w,y+h), (0,0,255), 2)
```

```python
import cv2
import numpy as np
import IPython.display as dp
from google.colab.patches import cv2_imshow
from pyzbar import pyzbar

p1=cv2.CascadeClassifier("haarcascades/haarcascade_frontalface_default.xml")

m1=cv2.imread("7.jpg", 1)
ret=p1.detectMultiScale(m1,minNeighbors=7,minSize=(10,10))
for x,y,w,h in ret:
  cv2.rectangle(m1, (x,y), (x+w,y+h), (0,0,255), 2)

cv2_imshow(m1)
```
